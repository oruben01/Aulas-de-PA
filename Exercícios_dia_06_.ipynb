```python
# 1 - Contagem simples: O usuário digita um número e o programa imprime do zero até o número.

# Entrada

num = int(input("Digite um número: "))

# Processamento

for i in range (0, num + 1):
  print(i)
```

```python
# 2 - Contagem regressiva: o usuário digita um número e o programa imprime do número até o zero.

# Entrada

num = int(input("Digite um número: "))
for i in range (num,- 1, -1):
  print(i)
```

```python
# 3 - Números pares: o usuário digita um número e o programa imprimi os números pares.

# Entrada

num = int(input("Digite um número: "))

# Processamento e Saída

for i in range (0, num + 1, 2):
  print(i)
```

```python
# 4 - Números ímpares: o usuário digita um número e o programa imprimi os números ímpares.

# Entrada

num = int(input("Digite um número: "))

# Processamento

for i in range (1, num + 1, 2):
  print(i)
```

```python
# 8 - Contar vogais: o usuário digita uma palavra e o programa conta as vogais.

# Entrada

palavra = str(input("Digite uma palavra: "))
vogais = "aeiou"
letra = 0

for i in range (len(palavra)):
  if palavra[i] in vogais:
    letra += 1

print(f"A palavra tem {letra} vogais.")
```
