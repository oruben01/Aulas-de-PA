```python
# Escrever um programa que o usuário digita uma palavra e o programa verifica se a palavra é uma palíndrome e imprime a palavra.

palavra = str(input("Digite uma palavra: "))
invertida = ""

# Entrada

for i in (palavra):
  invertida = i + invertida

if invertida == palavra:
  print(f"{palavra} {invertida}")
else:
  print(f"{palavra} {invertida}")



```

```python
# Soma acumulada
# Peça ao usuário para digitar um número e imprimi o resultado da soma de 1 até o número digitado

# Entrada

num = int(input("Digite um número:" ))

# Processamento

for i in range (0,num, ):
  print(i)
```

```python
# Peça um número ao usuário e o programa irá separar número pares e ímpares
# Exemplo de saída:

# Pares: 0, 2, 4, 6 ...
# Impares: 1, 3, 5 ...

# Entrada

num = int(input(f"Digite um número: "))

# Processamento

print("PARES")
for i in range (0,num, 2):
    print(i)

print("ÍMPARES")
for i in range (1,num, 2):
  print(i)
```
