```python
numero = int(input("Digite um número: "))

if numero % 2 == 0:
  print (f"O número {numero}, é um número par. ")
else:
  print (f"O número {numero}, é um número ímpar. ")
```

```python
# Escrever um programa que o usuário digita o nome e suas 04 notas referente aos bimestres, calcula a média e mostra
# aprovado, recuperação ou reprovado.
# Nota maior ou igual a 7 aprovado;
# Nota maior ou igual a 5 e menor que 7 recuperação;
#nota menor que 5 reprovado;

#Entrada

nome = str (input("Digite seu nome:"))
nota1 = float (input("Digite sua nota do 1º bimestre:"))
if nota1 > 10:
  print("Digite uma nota válida (notas de 0 a 10)")
nota2 = float (input("Digite sua nota do 2º bimestre:"))
if nota2 > 10:
  print("Digite uma nota válida (notas de 0 a 10)")
nota3 = float (input("Digite sua nota do 3º bimestre:"))
if nota3 > 10:
  print("Digite uma nota válida (notas de 0 a 10)")
nota4 = float (input("Digite sua nota do 4º bimestre:"))
if nota4 > 10:
  print("Digite uma nota válida (notas de 0 a 10)")

media = (nota1 + nota2 + nota3 + nota4) / 4

# Processamento


if media >= 7:
  print (f"A média de suas notas {nome}, foi {media}, você foi aprovado.")

elif media >= 5:
  print (f"A média de suas notas {nome}, foi {media}, você está de recuperação.")

else:
  print (f"A média de suas notas {nome}, foi {media}, você foi REPROVADO.")

# Saída
```

```python
# Entrada
lado1 = float(input("Digite o comprimento do primeiro lado:"))
lado2 = float(input("Digite o comprimento do segundo lado:"))
lado3 = float(input("Digite o comprimento do terceiro lado:"))

# Processamento e Saída

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
  print("É um triângulo válido!")
  if lado1 == lado2 and lado2 == lado3:
    print("É um triângulo Equilátero.")
  elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
    print("É um triângulo Isósceles.")
  else:
    print("É um triângulo Escaleno.")
else:
  print("Não é um triângulo válido.")
```
